```python
a = "змінна з текстом"
b = 1
c = ["a", 1, 1.25, "Слово", a]  # List (Список)
d = {"a": "Слово", "b": 1, a: b}  # Dict (Словник)
e = ("a", a)  # Tuple (незмінний список)
f = {"ss", a + str(b)}  # Set (Множина)
print("List:", c)
print("Dict:", d)
print("Set:", f)
```

```python
import keyword

words_to_check = ["hello", "class", "print", "continue"]

for word in words_to_check:
    is_reserved = keyword.iskeyword(word)
    print(f"Чи є слово '{word}' зарезервованим? -> {is_reserved}")
```

```python
scores = [5, 14, 8, 25, 3]
print(f"Максимальне значення (max): {max(scores)}")
print(f"Мінімальне значення (min): {min(scores)}")
print(f"Сума всіх чисел (sum): {sum(scores)}")
print(f"Кількість елементів (len): {len(scores)}")
```

```python
counter = 10
print("Запуск зворотного відліку...")

while counter > 0:
    print(counter)
    counter -= 1
    if counter == 3:
        print("Аварійна зупинка!")
        break
```

```python
import random

ages = [20, 12, 8]

for i in range(3):
    age_choice = random.choice(ages)
    print("Випав вік:", age_choice)

    if age_choice >= 18:
        print("Доступ дозволено")
    elif age_choice >= 12:
        print("Доступ обмежено")
    else:
        print("Доступ заборонено")
```

```python
user = {"name": "Rostik", "role": "admin"}

try:
    print(user["age"])
except KeyError as error:
    print(f"Помилка: Ключ {error} не знайдено в словнику.")
```

```python
words = ["apple", "cat", "banana", "dog", "elephant"]

short_words = list(filter(lambda w: len(w) < 5, words))

print(f"Короткі слова: {short_words}")
```
